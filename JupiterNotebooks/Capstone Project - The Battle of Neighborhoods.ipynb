## Problem 

We have a firm that want to expand and thus establish a new Coffee shop in either New York City or Toronto (Lets call them "Bad Morning Coffee" just for the fun of it). \
To do so they want to get an overview of how tough the competition is in each of the city as well as which district in the respective city that has the best environment for the shop. \
So the problem is "basically" just to find out which district how the right amount of coffee shops so that it is not a "dead" district but not an overfilled either. \
Lastly one could argue to look at the price of setling in the different places or cities as New York City is expensive compared to other places.
